```python
import pandas as pd
import numpy as np
from IPython.display import display
pd.options.display.max_columns = None
```

```python
df_sa_1 = pd.read_csv('./output_mushroom_data_1.csv')
df_sa_1.head()
```

```python
df_mushroom_labels = pd.read_csv("./mushroom_labels.csv")
df_mushroom_labels.head()
```

```python
df_join_shrooms = df_sa_1.merge(df_mushroom_labels, on="ID")
df_join_shrooms.head()
```

```python
T1 = pd.crosstab(df_join_shrooms.assignment,[df_join_shrooms.type]).round(2)
T1
```

```python
groupings = df_join_shrooms.drop(["ID","type"],axis=1).groupby("assignment").mean()
groupings.head()
```

```python
groupings = df_join_shrooms.drop(["ID","type"],axis=1).groupby("assignment").sum()
groupings.head()
```

```python
groupings = df_join_shrooms.drop(["ID","type"],axis=1).groupby("assignment").median()
groupings.head()
```
